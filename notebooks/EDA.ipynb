# Import Visualization Libraries
I will use `matplotlib` and `seaborn` to create visualizations for better understanding of fraudulent claims.


```python
import matplotlib.pyplot as plt
import seaborn as sns

# Set seaborn style for better visuals
sns.set_style("whitegrid")
```

# Count of Fraudulent vs. Non-Fraudulent Claims
This bar chart shows the number of fraudulent vs. non-fraudulent claims.


```python
import matplotlib.pyplot as plt
import seaborn as sns

# Set figure size
plt.figure(figsize=(6, 4))

# ðŸ”¹ **Added 'hue="fraud_reported"' to explicitly define the color grouping**
sns.countplot(data=df, x="fraud_reported", hue="fraud_reported", palette="coolwarm")

# Add labels and title
plt.xlabel("Fraudulent vs. Non-Fraudulent Claims")
plt.ylabel("Count")
plt.title("Count of Fraudulent vs. Non-Fraudulent Claims")

# Show plot
plt.show()
```

# Distribution of Total Claim Amounts
This histogram shows the spread of total claim amounts to detect unusual high-value claims.


```python
plt.figure(figsize=(8, 5))
sns.histplot(df["total_claim_amount"], bins=50, kde=True, color="blue")
plt.title("Distribution of Total Claim Amounts")
plt.xlabel("Claim Amount ($)")
plt.ylabel("Count")
plt.show()
```

# Boxplot of Claim Amounts
This boxplot will help identify extreme outliers in claim amounts.


```python
plt.figure(figsize=(8, 5))
sns.boxplot(x=df["total_claim_amount"], color="red")
plt.title("Boxplot of Claim Amounts")
plt.xlabel("Claim Amount ($)")
plt.show()
```

# Most Common Incident Types
This bar chart shows which types of insurance claims are most common.


```python
plt.figure(figsize=(8, 5))
sns.countplot(y=df["incident_type"], order=df["incident_type"].value_counts().index, palette="viridis")
plt.title("Most Common Incident Types")
plt.xlabel("Count")
plt.ylabel("Incident Type")
plt.show()
```
